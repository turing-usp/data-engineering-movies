## Extração e tratamento dos dados na API 

- Primeiro, é preciso definir os campos a mais que são desejados par ao dataset de filmes
- O campo de `imdb_id` é importante para buscar no crawler
- `revenue` e `budget` são campos que abrem espaço para diversas análises
- Para obter as `collections` e `productions` relacionadas a cada filme, foi usada uma função que obtém apenas o nome de cada um desses elementos listados dentro do filme

```python
movie_details = [
    'imdb_id', 
    'status', 
    'revenue', 
    'budget',
    {'fieldName': 'belongs_to_collection', 'finalName': 'collection', 'key': 'name'},
    {'fieldName': 'production_companies', 'finalName': 'productions', 'key': 'name'}
]
serie_details = detail_fields = [
    'in_production',
    'episode_run_time',
    'number_of_seasons',
    'number_of_seasons',
    'number_of_episodes',
    {'fieldName' : 'networks', 'finalName': 'networks', 'key': 'name'}
]
```

- Extranindo dados da rota trending para filmes: o _placeholder_ `:t` é substituído pelo tipo de mídia da extração (nesse caso, `movie`)

```python
trending_movies = get_route('/trending/:t/week', 'movie', movie_details)
```
